```
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```
N=500
k=[0.4,0.85,1,2,5]
```

```
for i in k:
    p=i/(N-1)
    G=nx.erdos_renyi_graph(N,p)
    nx.draw(G, pos=None, node_color='r', edge_color='b',node_size=5)
    plt.show()
```
